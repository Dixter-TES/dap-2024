## Ответы на вопросы

1. Типы ячеек в Jupyter.
Существует три основных типа: Code (для выполнения кода), Markdown (для форматированного текста) и Raw (для необработанного текста).

2. Как с помощью Markdown сформировать иерархический список?
Нужно сделать отступ (табуляция или 2-4 пробела) перед маркером вложенного элемента списка.

3. Как с помощью Markdown оформить цитату?
Необходимо поставить символ «>» (знак больше) перед текстом цитаты.

4. Как в текстовую ячейку вставить отформатированный блок кода?
Нужно заключить код в тройные обратные апострофы (```) сверху и снизу блока.

5. Сколько уровней заголовка существует в Markdown?
Существует 6 уровней заголовков, обозначаемых количеством символов «#» (от одного до шести).

6. Шаги процесса анализа данных, рассказать подробно про один из шагов.
Основные шаги: Постановка задачи, Сбор данных, Очистка данных, Разведочный анализ (EDA), Моделирование, Визуализация, Интерпретация результатов. Подробно про очистку данных: Это процесс обнаружения и исправления ошибок, пропусков, дубликатов и выбросов в «сырых» данных для приведения их к качественному виду, пригодному для анализа.

7. Типы данных, рассказать подробно про один из типов данных.
Основные типы: целые числа (int), числа с плавающей точкой (float), строки (str), списки (list), кортежи (tuple), словари (dict), множества (set), булевы значения (bool). Подробно про список (list): Это изменяемая упорядоченная коллекция произвольных объектов (чисел, строк и т.д.). Элементы в списке хранятся последовательно и к ним можно обращаться по индексу.

8. Что такое Jupyter Notebook, для чего используется?
Это интерактивная среда разработки, объединяющая код, текст, формулы и графики в одном документе. Используется для экспериментов с данными, обучения, визуализации и создания отчетов.

9. Почему python популярен для анализа данных?
Из-за простого синтаксиса, мощных библиотек для работы с данными (Pandas, NumPy, Matplotlib) и большого сообщества разработчиков.

10. Обязательно ли использовать Jupyter Notebook для анализа данных?
Нет, не обязательно. Можно писать код в обычных IDE (PyCharm, VS Code), но Jupyter удобнее для визуализации промежуточных результатов и исследовательского анализа.
